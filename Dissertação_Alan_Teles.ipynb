```python
import pandas as pd
from google.colab import drive
import datetime

# Montar Google Drive
drive.mount('/content/drive')

# Caminhos originais
arquivo1 = '/content/drive/MyDrive/MPES 20231 - Dissertação - Alan Teles/csv_parquet/RelatorioPE_anon_20250830_144720.parquet'

# Ler Parquet original
df1 = pd.read_parquet(arquivo1)
print("Antes da exclusão:")
display(df1.head())

# ----------------------
# Remover ocorrências indesejadas
# ----------------------
remover = ['0','5','G','I','Y']
df1 = df1[~df1['sg_ocorrencia'].isin(remover)].reset_index(drop=True)

print("Após exclusão das ocorrências:")
display(df1.head())

# ----------------------
# Salvar o DataFrame atualizado em Parquet
# ----------------------
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
arquivo_atualizado = f'/content/drive/MyDrive/MPES 20231 - Dissertação - Alan Teles/csv_parquet/RelatorioPE_anon_{timestamp}_atualizado.parquet'

df1.to_parquet(arquivo_atualizado, index=False)
print(f"Arquivo atualizado salvo em: {arquivo_atualizado}")

# ----------------------
# Substituir df1 pelo arquivo atualizado (para futuras execuções)
# ----------------------
df1 = pd.read_parquet(arquivo_atualizado)
print("df1 agora aponta para o Parquet atualizado.")
```

```python
import pandas as pd
from tqdm import tqdm
from datetime import datetime

# Timestamp para salvar arquivos
datastamp = datetime.now().strftime("%Y%m%d_%H%M%S")
save_path = "/content/drive/MyDrive/MPES 20231 - Dissertação - Alan Teles/csv_parquet/"

# ----------------------
# 1️ Normalizar datas e valores
# ----------------------
df2['DT_APONTAMENTO'] = pd.to_datetime(df2['DT_APONTAMENTO'], errors='coerce').dt.date
df2['VALOR_PROTESTO'] = pd.to_numeric(df2['VALOR_PROTESTO'].astype(str).str.strip(), errors='coerce')

df1['dt_protocolo'] = pd.to_datetime(df1['dt_protocolo'], errors='coerce').dt.date
df1['vl_protesto'] = pd.to_numeric(df1['vl_protesto'].astype(str).str.strip(), errors='coerce')

# ----------------------
# 2️ Criar chave de busca Documento + Cidade
# ----------------------
df2['CHAVE_BUSCA'] = df2['NUMERO_DOC_DEVEDOR'].astype(str) + '|' + df2['CIDADE'].str.strip().str.upper()
df1['CHAVE_BUSCA'] = df1['nm_nr_documento_devedor'].astype(str) + '|' + df1['nm_comarca'].str.strip().str.upper()

# ----------------------
# 3️ Sicase → Relatório
# ----------------------
rel_grp = df1.groupby('CHAVE_BUSCA')
matches_sic_rel, dismatches_sic_rel = [], []

for _, row in tqdm(df2.iterrows(), total=len(df2), desc="Sicase → Relatório"):
    chave = str(row['NUMERO_DOC_DEVEDOR']) + '|' + (row['CIDADE'].strip().upper() if pd.notna(row['CIDADE']) else '')
    candidatos = rel_grp.get_group(chave) if chave in rel_grp.groups else pd.DataFrame()

    match_doc = 'SIM' if not candidatos.empty else 'NÃO'
    if not candidatos.empty:
        match_data = 'SIM' if ((candidatos['dt_protocolo'] == row['DT_APONTAMENTO']).any()) else 'NÃO'
        match_valor = 'SIM' if ((candidatos['vl_protesto'] - row['VALOR_PROTESTO']).abs() < 0.01).any() else 'NÃO'
    else:
        match_data = 'NÃO'
        match_valor = 'NÃO'

    resultado = row.to_dict()
    resultado.update({'MATCH_DOCUMENTO': match_doc, 'MATCH_DATA': match_data, 'MATCH_VALOR': match_valor})

    if all([match_doc=='SIM', match_data=='SIM', match_valor=='SIM']):
        matches_sic_rel.append(resultado)
    else:
        dismatches_sic_rel.append(resultado)

# ----------------------
# 4️ Relatório → Sicase
# ----------------------
sic_grp = df2.groupby('CHAVE_BUSCA')
matches_rel_sic, dismatches_rel_sic = [], []

for _, row in tqdm(df1.iterrows(), total=len(df1), desc="Relatório → Sicase"):
    chave = str(row['nm_nr_documento_devedor']) + '|' + (row['nm_comarca'].strip().upper() if pd.notna(row['nm_comarca']) else '')
    candidatos = sic_grp.get_group(chave) if chave in sic_grp.groups else pd.DataFrame()

    match_doc = 'SIM' if not candidatos.empty else 'NÃO'
    if not candidatos.empty:
        match_data = 'SIM' if ((candidatos['DT_APONTAMENTO'] == row['dt_protocolo']).any()) else 'NÃO'
        match_valor = 'SIM' if ((candidatos['VALOR_PROTESTO'] - row['vl_protesto']).abs() < 0.01).any() else 'NÃO'
    else:
        match_data = 'NÃO'
        match_valor = 'NÃO'

    resultado = row.to_dict()
    resultado.update({'MATCH_DOCUMENTO': match_doc, 'MATCH_DATA': match_data, 'MATCH_VALOR': match_valor})

    if all([match_doc=='SIM', match_data=='SIM', match_valor=='SIM']):
        matches_rel_sic.append(resultado)
    else:
        dismatches_rel_sic.append(resultado)

# ----------------------
# 5️ Criar DataFrames finais
# ----------------------
df_matches_sic_rel = pd.DataFrame(matches_sic_rel)
df_dismatches_sic_rel = pd.DataFrame(dismatches_sic_rel)
df_matches_rel_sic = pd.DataFrame(matches_rel_sic)
df_dismatches_rel_sic = pd.DataFrame(dismatches_rel_sic)

# ----------------------
# 6️ Função de estatísticas gerais
# ----------------------
def estatisticas(matches_df, dismatches_df, nome=""):
    total = len(matches_df) + len(dismatches_df)
    print(f"\n📊 Estatísticas: {nome}")
    print(f"Total de registros: {total}")
    print(f"Matches: {len(matches_df)} ({len(matches_df)/total*100:.2f}%)")
    print(f"Dismatches: {len(dismatches_df)} ({len(dismatches_df)/total*100:.2f}%)")

estatisticas(df_matches_sic_rel, df_dismatches_sic_rel, "Sicase → Relatório")
estatisticas(df_matches_rel_sic, df_dismatches_rel_sic, "Relatório → Sicase")

# ----------------------
# 7️ Estatísticas por município
# ----------------------
def estatisticas_por_municipio(df, tipo=""):
    col_cidade = 'CIDADE' if 'CIDADE' in df.columns else 'nm_comarca'
    df['MATCH_TOTAL'] = df[['MATCH_DOCUMENTO','MATCH_DATA','MATCH_VALOR']].apply(lambda x: all(x=='SIM'), axis=1)
    resumo = df.groupby(col_cidade)['MATCH_TOTAL'].agg(['sum','count'])
    resumo['percentual'] = resumo['sum']/resumo['count']*100
    print(f"\n📍 Estatísticas por município ({tipo})")
    display(resumo.sort_values('percentual', ascending=False).head(20))

estatisticas_por_municipio(pd.concat([df_matches_sic_rel, df_dismatches_sic_rel]), "Sicase → Relatório")
estatisticas_por_municipio(pd.concat([df_matches_rel_sic, df_dismatches_rel_sic]), "Relatório → Sicase")

# ----------------------
# 8️ Estatísticas por situação do título
# ----------------------
def estatisticas_por_situacao(df, tipo=""):
    if 'sg_ocorrencia' in df.columns:
        df['MATCH_TOTAL'] = df[['MATCH_DOCUMENTO','MATCH_DATA','MATCH_VALOR']].apply(lambda x: all(x=='SIM'), axis=1)
        resumo = df.groupby('sg_ocorrencia')['MATCH_TOTAL'].agg(['sum','count'])
        resumo['percentual'] = resumo['sum']/resumo['count']*100
        print(f"\n📌 Estatísticas por situação do título ({tipo})")
        display(resumo.sort_values('percentual', ascending=False))

estatisticas_por_situacao(pd.concat([df_matches_sic_rel, df_dismatches_sic_rel]), "Sicase → Relatório")
estatisticas_por_situacao(pd.concat([df_matches_rel_sic, df_dismatches_rel_sic]), "Relatório → Sicase")

# ----------------------
# 9️ Motivos dos dismatches
# ----------------------
def motivos_dismatches(df_dismatches, tipo=""):
    df_dismatches['MOTIVO'] = df_dismatches.apply(
        lambda x: ", ".join([
            'Documento não encontrado' if x['MATCH_DOCUMENTO']=='NÃO' else '',
            'Data divergente' if x['MATCH_DATA']=='NÃO' else '',
            'Valor divergente' if x['MATCH_VALOR']=='NÃO' else ''
        ]).strip(', '),
        axis=1
    )
    print(f"\n⚠️ Estatísticas de dismatches ({tipo}):")
    resumo = df_dismatches['MOTIVO'].value_counts().to_frame()
    resumo.columns = ['quantidade']
    resumo['percentual'] = resumo['quantidade']/resumo['quantidade'].sum()*100
    display(resumo)
    return df_dismatches

df_dismatches_sic_rel = motivos_dismatches(df_dismatches_sic_rel, "Sicase → Relatório")
df_dismatches_rel_sic = motivos_dismatches(df_dismatches_rel_sic, "Relatório → Sicase")

# ----------------------
# 10 Motivos por município
# ----------------------
def motivos_por_municipio(df_dismatches, tipo=""):
    col_cidade = 'CIDADE' if 'CIDADE' in df_dismatches.columns else 'nm_comarca'
    resumo = df_dismatches.groupby([col_cidade, 'MOTIVO']).size().unstack(fill_value=0)
    print(f"\n📍 Motivos de dismatches por município ({tipo})")
    display(resumo)

motivos_por_municipio(df_dismatches_sic_rel, "Sicase → Relatório")
motivos_por_municipio(df_dismatches_rel_sic, "Relatório → Sicase")

# ----------------------
# 11 Motivos por situação do título
# ----------------------
def motivos_por_situacao(df_dismatches, tipo=""):
    if 'sg_ocorrencia' in df_dismatches.columns:
        resumo = df_dismatches.groupby(['sg_ocorrencia','MOTIVO']).size().unstack(fill_value=0)
        print(f"\n📌 Motivos de dismatches por situação do título ({tipo})")
        display(resumo)

motivos_por_situacao(df_dismatches_sic_rel, "Sicase → Relatório")
motivos_por_situacao(df_dismatches_rel_sic, "Relatório → Sicase")

# ----------------------
# 12 Salvar DataFrames em Parquet
# ----------------------
df_matches_sic_rel.to_parquet(f"{save_path}MatchesSicaseRelatorio_{datastamp}.parquet", index=False)
df_dismatches_sic_rel.to_parquet(f"{save_path}DismatchesSicaseRelatorio_{datastamp}.parquet", index=False)
df_matches_rel_sic.to_parquet(f"{save_path}MatchesRelatorioSicase_{datastamp}.parquet", index=False)
df_dismatches_rel_sic.to_parquet(f"{save_path}DismatchesRelatorioSicase_{datastamp}.parquet", index=False)

print("\n Processamento completo. Todos os arquivos foram salvos no Google Drive!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Configurações gerais de estilo
sns.set(style="whitegrid")
plt.rcParams.update({'figure.autolayout': True, 'font.size': 12})

# ----------------------
# 1️ Distribuição geral de matches vs dismatches
# ----------------------
total = len(df_matches_sic_rel) + len(df_dismatches_sic_rel)
labels = ['Matches', 'Dismatches']
sizes = [len(df_matches_sic_rel)/total*100, len(df_dismatches_sic_rel)/total*100]
colors = ['#4CAF50','#F44336']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize':12})
plt.title("Distribuição de Matches x Dismatches (Sicase → Relatório)")
plt.savefig(f"{save_path}Distribuicao_Matches_Dismatches_{datastamp}.png", dpi=300)
plt.show()

# ----------------------
# 2️ Percentual de matches por município
# ----------------------
resumo_mun = pd.concat([df_matches_sic_rel, df_dismatches_sic_rel])
resumo_mun['MATCH_TOTAL'] = resumo_mun[['MATCH_DOCUMENTO','MATCH_DATA','MATCH_VALOR']].apply(lambda x: all(x=='SIM'), axis=1)
resumo_mun = resumo_mun.groupby('CIDADE')['MATCH_TOTAL'].agg(['sum','count'])
resumo_mun['percentual'] = resumo_mun['sum']/resumo_mun['count']*100
resumo_mun = resumo_mun.sort_values('percentual', ascending=True)

plt.figure(figsize=(10,8))
sns.barplot(x='percentual', y=resumo_mun.index, data=resumo_mun, palette='Blues_r')
plt.xlabel('Percentual de Matches (%)')
plt.ylabel('Município')
plt.title('Percentual de Matches por Município')
plt.savefig(f"{save_path}Matches_por_Municipio_{datastamp}.png", dpi=300)
plt.show()

# ----------------------
# 3️ Frequência de motivos dos dismatches
# ----------------------
resumo_motivos = df_dismatches_sic_rel['MOTIVO'].value_counts().reset_index()
resumo_motivos.columns = ['Motivo','Quantidade']

plt.figure(figsize=(8,6))
sns.barplot(x='Quantidade', y='Motivo', data=resumo_motivos, palette='Oranges_r')
plt.title('Motivos de Dismatches (Sicase → Relatório)')
plt.xlabel('Quantidade')
plt.ylabel('Motivo')
plt.savefig(f"{save_path}Motivos_Dismatches_{datastamp}.png", dpi=300)
plt.show()

# ----------------------
# 4️ Motivos de dismatches por município (stacked)
# ----------------------
resumo_stack = df_dismatches_sic_rel.groupby(['CIDADE','MOTIVO']).size().unstack(fill_value=0)
resumo_stack = resumo_stack.sort_values(resumo_stack.columns[0], ascending=True)  # ordena por um motivo para visualização

resumo_stack.plot(kind='barh', stacked=True, figsize=(12,8), colormap='tab20')
plt.xlabel('Quantidade de Dismatches')
plt.ylabel('Município')
plt.title('Motivos de Dismatches por Município')
plt.legend(title='Motivo', bbox_to_anchor=(1.05,1), loc='upper left')
plt.savefig(f"{save_path}Motivos_por_Municipio_{datastamp}.png", dpi=300, bbox_inches='tight')
plt.show()
```
